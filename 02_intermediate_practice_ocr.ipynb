## Importing modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Thresholding

#### A. Binary Thresholding

```python
image = cv2.imread('images/digits.jpg', cv2.IMREAD_GRAYSCALE)
T=127 #  threshold value
_, thresh1 = cv2.threshold(image, T, 255, cv2.THRESH_BINARY)
_, thresh2 = cv2.threshold(image, T, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(16, 9))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Binary Thresholding Image")
plt.axis("off")
plt.imshow(thresh1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Binary Inverse Thresholding Image")
plt.axis("off")
plt.imshow(thresh2, cmap='gray')
```

#### B. Otsu's Thresholding

```python
T1, otsu1 = cv2.threshold(image, thresh=0, maxval=255, type=cv2.THRESH_BINARY+cv2.THRESH_OTSU)
T2, otsu2 = cv2.threshold(image, thresh=0, maxval=255, type=cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

print(f"The dynamic T1 of Otsu's Thresholding is: {T1}")
print(f"The dynamic T2 of Otsu's Thresholding is: {T2}")

plt.figure(figsize=(16, 9))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Otsu's Thresholding Image")
plt.axis("off")
plt.imshow(otsu1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Otsu's Inverse Thresholding Image")
plt.axis("off")
plt.imshow(otsu2, cmap='gray')
```

#### C. Otsu's Thresholding with Gaussian Blur

```python
gb = cv2.GaussianBlur(image, ksize=(5,5), sigmaX=0)

t1, o1 = cv2.threshold(src=gb, thresh=0, maxval=255, type=cv2.THRESH_BINARY+cv2.THRESH_OTSU)
t2, o2 = cv2.threshold(src=gb, thresh=0, maxval=255, type=cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

plt.figure(figsize=(16, 9))
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(image, cmap='gray')

plt.subplot(2, 2, 2)
plt.title("Gaussian Blur Image")
plt.axis("off")
plt.imshow(gb, cmap='gray')

plt.subplot(2, 2, 3)
plt.title("Otsu's Thresholding Image (with Gaussian Blur)")
plt.axis("off")
plt.imshow(o1, cmap='gray')

plt.subplot(2, 2, 4)
plt.title("Otsu's Inverse Thresholding Image (with Gaussian Blur)")
plt.axis("off")
plt.imshow(o2, cmap='gray')
```
