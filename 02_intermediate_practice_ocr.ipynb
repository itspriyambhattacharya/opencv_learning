## Importing modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Thresholding

#### A. Binary Thresholding

```python
image = cv2.imread('images/digits.jpg', cv2.IMREAD_GRAYSCALE)
T=120 #  threshold value
retval1, thresh1 = cv2.threshold(image, T, 255, cv2.THRESH_BINARY)
retval2, thresh2 = cv2.threshold(image, T, 255, cv2.THRESH_BINARY_INV)

print(f"Retval 1 is: {retval1} and retval 2 is: {retval2}")

plt.figure(figsize=(16, 9))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Binary Thresholding Image")
plt.axis("off")
plt.imshow(thresh1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Binary Inverse Thresholding Image")
plt.axis("off")
plt.imshow(thresh2, cmap='gray')
```
